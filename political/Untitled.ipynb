```python
def clean(my_dict):
    try:
        for k in list(my_dict.keys()):
            if my_dict[k] in ['',{},None]:
                del my_dict[k]
            else:
                my_dict[k]=clean(my_dict[k])
        return my_dict
    except AttributeError:
        return my_dict
        
```
